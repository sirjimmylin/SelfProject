```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact

# Load the data
data = pd.read_csv('/Users/jimmylin/Documents/GitHub/Own Environment/Motor_Vehicle_Collisions_-_Crashes_20250107.csv')

# Convert 'CRASH TIME' to datetime if not already
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Function to plot top contributing factors for a given hour
def plot_top_factors(hour):
    # Filter data for the given hour
    filtered_data = data[data['CRASH TIME'].dt.hour == hour]
    top_factors = filtered_data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(10)

    # Clear the current figure
    plt.clf()

    # Plot the top contributing factors
    plt.figure(figsize=(12, 7))
    sns.barplot(x=top_factors.index, y=top_factors.values, hue=top_factors.index, palette="magma", legend=False)
    plt.title(f'Top 10 Contributing Factors to Crashes at {hour:02d}:00', fontsize=16)
    plt.xlabel('Top Factors', fontsize=14)
    plt.ylabel('Number of Crashes', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Create an interactive slider for the hour
interact(plot_top_factors, hour=(0, 23));
```
